# Работа с типом мероприятий </br> 

```python
def check_type(frame):
    event_types = ['Вебинар', 'Семинар', 'Мастер-класс','День открытых дверей','Конференция','Консультация','Презентация','Тренинг','Конкурс','Дискуссия','Хакатон','Встреча','Ярмарка вакансий','Воркшоп', 'Кейс-чемпионат', 'Лекция']
    for i in event_types:
        if i.lower()[1:] in str(frame).lower():
            return i
    return 'Другое'
```

```python
frame['Тип мероприятия'] = frame['Наименование мероприятия'].apply(check_type)
```

```python
frame['Тип мероприятия'].value_counts()
```

```python
frame[frame['Тип мероприятия'] == 'Другое']
```

```python
data = list(frame[frame['Тип мероприятия'] == 'Другое']['Наименование мероприятия'])
```

```python
def change_sym(string):
    string = str(string).lower()
    string = string.replace('.', '')
    string = string.replace(',', '')
    string = string.replace(':', '')
    string = string.replace(';', '')
    string = string.replace('»', '')
    string = string.replace('«', '')
    string = string.replace('&', '')
    string = string.replace('?', '')
    string = string.replace('"', '')
    string = string.replace("'", '')
    string = string.replace("“", '')
    string = string.replace("”", '')
    string = string.replace("”", '')
    string = string.replace("*", '')
    string = string.replace("!", '')
    string = string.replace("!", '')
    string = string.replace(")", '')
    string = string.replace("(", '')
    string = string.replace("`", '')
    string = string.replace("\xa0", '')
    return string

new_sentences = []
for i in data:
    i = change_sym(str(i))
    new_sentences.append(i)

all_words = []
for i in new_sentences:
    temp = i.split()
    for i in temp:
        all_words.append(i)

words_counter = {}
for i in all_words:
    if i not in words_counter:
        words_counter[i] = 1
    else:
        words_counter[i] +=1

words_counter = {k: words_counter[k] for k in sorted(words_counter, key=words_counter.get, reverse=True)}
words_counter
```

# Работа с местом

```python
def check_place(place):
    place = str(place).lower()
    if 'шабол' in place or 'кирпич' in place:
        return'Шаболовская'
    elif 'мясниц' in place:
        return 'Мясницкая'
    elif 'онлайн' in place or 'online' in place or 'zoom' in place or 'www' in place or 'http' in place:
        return 'Онлайн'
    elif 'трифон' in place:
        return 'Трифоновская'
    else:
        return 'Отсутствует'
```

```python
frame['Место проведения кратко'] = frame['Место проведения'].apply(check_place)
```

```python
frame['Место проведения кратко'].value_counts()
```

```python
def add_place_according_to_name(name):
    name_1 = str(name).lower()
    if 'онлайн' in name_1 or 'online' in name_1 or 'zoom' in name_1 or 'www' in name_1 or 'http' in name_1:
        return 'Онлайн'
    else:
        return 'Отсутствует'
```

```python
frame.iloc[list(frame[frame['Место проведения кратко'] == 'Отсутствует'].index), -1] = frame.iloc[list(frame[frame['Место проведения кратко'] == 'Отсутствует'].index), 1].apply(add_place_according_to_name)
```
